```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#setando tamanho dos plots 
sb.set(rc={'figure.figsize':(15,8)})
```

A base de dados deste exercício é a "Quantidade de alunos por tipo de ensino" da rede estadual de São Paulo de 2021. 
Os dados estão sendo importados diretamente do [Portal de Dados Abertos da Educação do Estado de São Paulo](https://dados.educacao.sp.gov.br/dataset/quantidade-de-alunos-por-tipo-de-ensino-da-rede-estadual).
O dicionário de dados está disponível neste [link](https://dados.educacao.sp.gov.br/dicion%C3%A1rio-de-dados-de-quantidade-de-alunos-por-tipo-de-ensino)

```python
#importando dados
url = 'https://dados.educacao.sp.gov.br/sites/default/files/Quantidade%20de%20alunos%20por%20tipo%20de%20ensino%20da%20rede%20estadual%20-%202021.csv'
escolas = pd.read_csv(url)
```

```python
escolas.head(30)
```

```python
escolas.info()
```

```python
#Explorando alguns atributos com a função displot
sb.displot(escolas['ANOS INICIAIS'],bins=10)
plt.show()
```

Pelo gráfico notamos que muitas escolas não possuem ensino para os anos iniciais do "Ensino Fundamental", portanto possuem 0 nessa coluna.

```python
#Para facilitar a visualização vamos criar um filtro que mostra apenas as escolas que possuem alunos no anos iniciais do Ensino Fundamental
#Esse coluna possui o numero de alunos, portanto selecionaremos aquela coluna com valor acima de zero 
filtro_esc_fund = escolas[escolas['ANOS INICIAIS']>0]
sb.displot(filtro_esc_fund['ANOS INICIAIS'],bins=20)
plt.show()
```

```python
sb.displot(escolas['ANOS FINAIS'], bins=10)
plt.show;
```

```python
#Vamos analisar apenas as escolas que possuem os anos finais do ensino fundamental
filtro_esc_fund1 = escolas[escolas['ANOS FINAIS']>0]
sb.displot(filtro_esc_fund['ANOS FINAIS'],bins=20)
plt.show()
```

```python
#Agora analisaremos as escolas com ensino médio
sb.displot(escolas['ENSINO MEDIO'],bins=10)
plt.show;
```

```python
filtro_ens_medio = escolas[escolas['ENSINO MEDIO']>0]
sb.displot(filtro_esc_fund['ENSINO MEDIO'],bins=20)
plt.show()
```

Podemos observar que nos três casos analisados, a maior concentração de alunos está na faixa de 250 a 500 alunos.

```python
#Agora vamos analisar a proporção de alunos em cada fase de ensino.
fund_iniciais = escolas['ANOS INICIAIS'].sum()
fund_finais = escolas['ANOS FINAIS'].sum()
ensino_medio = escolas['ENSINO MEDIO'].sum()

print('Fundamental anos iniciais: ', fund_iniciais,
      '\nFundamental anos finais: ',fund_finais,
      '\nEnsino Médio:', ensino_medio)
```

```python
alunos = [fund_iniciais, fund_finais, ensino_medio]
periodo = ['Fundamental Inicial\n1º a 5º ano', 
           'Fundamental Final\n6º a 9º ano',
           'Ensino Médio']
```

```python
plt.pie(alunos, labels=periodo, autopct = '%0.0f%%')
plt.show()
```

Agora podemos observar que a menor proporção de alunos está no ensino fundamental inicial 

```python
#Agora vamos usar um gráfico boxplot para analisar a quantidade de alunos do ensino médio nas escolas em geral
sb.boxplot(y=filtro_ens_medio['ENSINO MEDIO'])
plt.show()
```

Podemos ver que a maioria das escolas tem entre 200 e 500 alunos do ensino médio

```python
escolas['ENSINO MEDIO'].describe()
```

```python
#Vamos utilizar um scatter plot para analisar visualmente se há correlação entre a quantidade de alunos de cada fase.
#Vamos iniciar pelo ensino fundamental final e ensino médio por terem mais alunos
filtro_fund2_medio = escolas[(escolas['ANOS FINAIS'] > 0) & (escolas['ENSINO MEDIO'] > 0)]
sb.scatterplot(x=filtro_fund2_medio['ANOS FINAIS'],y=filtro_fund2_medio['ENSINO MEDIO'])
plt.show()
```

```python
filtro_fund1_medio = escolas[(escolas['ANOS INICIAIS'] > 0) & (escolas['ENSINO MEDIO'] > 0)]
sb.scatterplot(x=filtro_fund1_medio['ANOS INICIAIS'],y=filtro_fund1_medio['ENSINO MEDIO'])
plt.show()
```

```python
filtro_funds = escolas[(escolas['ANOS INICIAIS'] > 0) & (escolas['ANOS FINAIS'] > 0)]
sb.scatterplot(x=filtro_funds['ANOS INICIAIS'],y=filtro_funds['ANOS FINAIS'])
plt.show()
```

Embora haja pontos em dipersão nos gráficos podemos notar que há correlação entre o número de alunos do ensino fundamental inicial (alunos de 1º a 5º ano) e ensino fundamental final (6º a 9º ano). Também pudemos observar correlação entre a quantidade de alunos do ensino fundamental final e ensino médio.
